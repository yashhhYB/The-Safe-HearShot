```python
from json import loads
with open("log.txt", "r") as f:
    lines = [loads(x.strip()) for x in f.readlines()]
    
base = []
for l in lines:
    base += l
```

```python
avg = [sum(l)/len(l) for l in lines]
max_ = [max(l) for l in lines]
```

```python
len(base)
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Plot the data
plt.plot(avg)
plt.xlabel('Time')
plt.ylabel('Avg Activation')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Plot the data
plt.plot(max_)
plt.xlabel('Time')
plt.ylabel('Max Activation')
plt.show()
```
